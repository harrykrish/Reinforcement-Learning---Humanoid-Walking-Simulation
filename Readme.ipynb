### Method - 3 - Exploitation using RGB- D Sensor (Kinect) and Exploration using reinforcement learning (DDPG)



## Steps
* Capture markers captured via BodyBasics WPF
* Associate markers to appropriate joints and bones in OpenSim
* Save the associate markers as .trc file
* Save the associated motion captured as .mot file.
* Feed the .trc file to the model and save that as one of the policies before training the model with reinforcement learning
